### Data cleaning e feature engineering

```python
import pandas as pd
```

```python
base = pd.read_csv('best-selling game consoles.csv')
```

```python
base.info()
```

```python
base.head()
```

```python
#vendo os comentários
base.loc[base.Remarks.notnull()]
```

```python
#nada muito relevante para nossa análise
base = base.drop('Remarks', axis =1 )
```

```python
#alterar o nome de algumas colunas
base = base.rename(columns={'Units sold (million)': 'sales', 'Released Year': 'Released_Year', 'Discontinuation Year': 'Discontinuation_Year'})
```

```python
#vendo tempo de duração
duration = base['Discontinuation_Year'] - base['Released_Year']
base['duration'] = duration
base.head()	
```

```python
base.loc[base['duration'] < 0]
#vemos que o item 44 tem algo estranho os demais apenas ainda estão lançando
#pesquisando na internet realmente está errado vamos ter que alterar
```

```python
#alterando para dados achados na internet
base.loc[base['Console Name'] == 'Magnavox Odyssey', 'Released_Year'] = 1975
base.loc[base['Console Name'] == 'Magnavox Odyssey', 'Discontinuation_Year'] = 1992
base.loc[base['Console Name'] == 'Magnavox Odyssey']
```

```python
#vou usar 2023 para mostrar que os jogos ainda continuam em produção
base.loc[base['Discontinuation_Year'] == 0, 'Discontinuation_Year'] = 2023
```

### Análise exploratória


```python
#vamos ver qual empresa vendeu mais
base.groupby(['Company'])['sales'].sum().sort_values(ascending =False)
```

```python
base.groupby(['Company'])['duration'].mean().sort_values(ascending =False)
```

```python
base.groupby(['Type'])['sales'].sum().sort_values(ascending =False)
```
